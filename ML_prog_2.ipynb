```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
print("Prediction for X=6:", model.predict([[6]]))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Input features (Area, Bedrooms)
X = np.array([
    [800, 2],
    [1000, 2],
    [1200, 3],
    [1500, 3],
    [1800, 4]
])

# Output
y = np.array([40, 50, 65, 80, 95])

model = LinearRegression()
model.fit(X, y)

# Prediction
print("Predicted price:",
      model.predict([[1400, 3]]))
```
